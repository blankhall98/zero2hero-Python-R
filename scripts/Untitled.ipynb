```python
INGRESOS = 25000
PERIODO = 12

gasto_fijo = 5000
gasto_variable = (INGRESOS-gasto_fijo)/4

EGRESOS = gasto_variable + gasto_fijo
```

```python
activo = False
if INGRESOS > EGRESOS:
    activo = True
```

```python
if activo:
    ganancia_anual = INGRESOS*PERIODO
    gasto_anual = EGRESOS*PERIODO
    
    ahorro = ganancia_anual - gasto_anual
    print(ahorro)
```

```python
#DATOS
COSTO_TERRENO = {
    'CDMX': 28000,
    'MTY': 21000,
    'GDL': 20000
}

COSTO_CONST = {
    'Residencial': (7000 + 11000)/2,
    'Economico': (5500 + 7500)/2
}

#VARIABLES DE DESICION
dimension_casa = 60
ciudad = 'CDMX'
modelo = 'Residencial'
```

```python
ct = COSTO_TERRENO[ciudad]*dimension_casa
cc = COSTO_CONST[modelo]*dimension_casa

COSTO_TOTAL = ct + cc
print(f'Costo de mi casa deseada: {COSTO_TOTAL}')
```

## Estrategia

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
T = 15

#cetes
simu_cete = [ahorro]
tasa = 11.1/100
for i in range(T):
    y = ahorro + simu_cete[i]*(1+tasa)
    simu_cete.append(y)

tiempo = np.arange(T+1)
simu = ahorro*tiempo

plt.figure(figsize = (10,6))

plt.title('estrategia para lograr mis metas')
plt.plot(tiempo,simu,label='ESTRATEGIA ESTACIONARIA')
plt.plot(6,COSTO_TOTAL,'ro',label='costo casa')
plt.plot(tiempo,simu_cete,label='ESTRATEGIA CETE')
plt.legend()
plt.show()
```
